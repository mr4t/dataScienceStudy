```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yas=np.linspace(10,70,13)
```

```python
kilo=np.random.randint(20,90,13)
```

```python
plt.plot(yas,kilo,"r") # x ekseni yas, y ekseni kilo, grafiğin rengi "r"
plt.xlabel("yas") # x ekseni ismi
plt.ylabel("kilo") # y ekseni ismi
plt.title("yasa göre kilo değişim grafiği") #grafik ismi
#plt.show()
```

```python
dizi1=np.linspace(0,10,20)
```

```python
dizi2=dizi1**3
```

```python
plt.plot(dizi1,dizi2,"b--")
```

### plt.subplot(1,2,x) subplot ekrana çizilen grafik sayısını belirler yanda 1 satır 2 sütundan oluşan grafik belirtilmiş x yerine çizilecek grafiğin konum bilgisi verilebilir(mesela 1 veya 2)

```python
plt.subplot(1,2,1)
plt.plot(dizi1,dizi2,"b--")
plt.subplot(1,2,2)
plt.plot(dizi2,dizi1,"g.-")
```

```python
plt.subplot(2,2,1)
plt.plot(dizi1,dizi2,"b--")
plt.subplot(2,2,2)
plt.plot(dizi1,dizi2,"b*-")
plt.subplot(2,2,3)
plt.plot(dizi1,dizi2,"b.-")
plt.subplot(2,2,4)
plt.plot(dizi1,dizi2,"b")
```

## matplotlib figure

```python
myFigure=plt.figure()
myAxes=myFigure.add_axes([0.2,0.2,0.9,0.9]) #listenin ilk iki elemanı grafiğin çizildiği pencerenin alt köşesine olan 
#uzaklık değeri, listenin son iki elemanı grafiğin height, witdh değerleri
myAxes.plot(dizi1,dizi2,"b")
myAxes.set_xlabel("name of the x axes")
myAxes.set_ylabel("name of the y axes")
myAxes.set_title("name of the title")
```

#### matplotlib figurleri iç içe grafik çizdirmek için kullanılabilir.

```python
figure1=plt.figure()
eksen1=figure1.add_axes([.1,.1,.9,.9])
eksen2=figure1.add_axes([.2,.5,.4,.4])

eksen1.plot(dizi1,dizi2,"b")
eksen1.set_xlabel("x axis")
eksen1.set_ylabel("y axis")
eksen1.set_title("Main Graphic")
eksen2.plot(dizi2,dizi1,"r")
eksen1.set_xlabel("small x axis")
eksen1.set_ylabel("small y axis")
eksen1.set_title("Small Graphic")
```

```python
(benimFigure,benimEksen)=plt.subplots(1,2)
benimEksen[0].plot(dizi1,dizi2,"b")
benimEksen[1].plot(dizi1,dizi2,"b")
#1 satır 2 sütunluk grafiği liste olarak alıp çizdirmlerini verdik
plt.tight_layout() #bu fonksiyon grafkler arasındaki açıklığı uygun hale getirir
benimFigure #çizilen her eksen bir pencere gibi davranan benimFigure içinde toplanır
```
