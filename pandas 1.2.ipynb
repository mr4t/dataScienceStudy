```python
import pandas as pd
import numpy as np
```

## Concatenation

```python
dic1={"isim":["ali","veli","güllü","banu","irem","burak"],
     "sinif":["A","B","C","D","E","F"],
     "yas":["12","23","24","22","15","19"]}
```

```python
dic2={"isim":["yaren","ipek","murat","can","ela","sema"],
     "sinif":["A","B","C","D","E","F"],
     "yas":["14","21","22","22","15","19"]}
```

```python
dic3={"isim":["alya","samet","güven","banu","irem","burak"],
     "sinif":["A","B","C","D","E","F"],
     "yas":["12","23","24","22","15","19"]}
```

```python
kisiDF1=pd.DataFrame(dic1,np.arange(1,7))
kisiDF2=pd.DataFrame(dic2,np.arange(7,13))
kisiDF3=pd.DataFrame(dic3,np.arange(13,19))
kisiDF4=pd.DataFrame(dic2,np.arange(1,7))
```

```python
kisiDF1
```

```python
kisiDF2
```

```python
kisiDF3
```

```python
pd.concat([kisiDF1,kisiDF2,kisiDF3]) #dataFrameleri alt alta birleştirir.
```

```python
pd.concat([kisiDF1,kisiDF2,kisiDF3,kisiDF4],axis=1).fillna("xx") #axis=1 yan yana birleştirme yapar kisiDF1 ile kisiDF4 ün 
#index bilgisi aynı olduğu için yana yana gelirken index bilgisi farklı olanda yan tarafa gelirk ve aynı zamanda 
#indexine uygun şekilde aşağı kayar
```

## merge (kaynaştırma)

```python
mergeDic1={"isim":["murat","ceren","haluk","hale"],
         "spor":["basketbol","futbol","tenis","voleybol"],
          "A":["basketbol","futbol","tenis","voleybol"],
          "B":["basketbol","futbol","tenis","voleybol"]}
mergeDic2={"isim":["murat","ceren","haluk","hale"],
         "kalori":["100","200","300","400"],
           "A":["100","200","300","400"],
          "B":["basketbol","futbol","tenis","voleybol"]}
```

```python
mergeDF1=pd.DataFrame(mergeDic1,[1,2,3,4])
mergeDF2=pd.DataFrame(mergeDic2,[1,2,3,4])
```

```python
pd.merge(mergeDF1,mergeDF2,on=["isim","B"]) #merge işlemi "on=..." diye belirtilen kolonu ortak alarak diğer kolonları tek bir
#DataFrame altında toplar (belirtilmemiş olan aynı isimli kolonlar x,y diye adlandırılır)

#ortak kolonları yukarıda liste olarak verdik ve isim ile B kolonunda aynı olan elemanlar için merge işlemi yapıldı eğer
#ortak alınan kolonlarda aynı olmayan eleman olsaydı elemanın denk geldiği satır silinirdi
```

## pandas ileri operasyonlar

```python
sinifDic={"isim":["ali","veli","güllü","banu","irem","burak"],
     "sinif":["A","B","C","D","C","F"],
     "yas":[12,23,24,22,15,19]}
```

```python
sinifDF=pd.DataFrame(sinifDic)
```

```python
sinifDF
```

```python
sinifDF["sinif"].unique() #sınıf içindeki herşeyi bir defa alır
```

```python
sinifDF["sinif"].nunique() #sınıf içinde bir defa alınan şeylerin toplam sayısını verir
```

```python
sinifDF["sinif"].value_counts() #sinif içindeki elemanların sayısını verir
```

```python
def yarimYas(x):
    return x/2
```

```python
sinifDF["yas"].apply(yarimYas) #yas kolonundaki her eleman için yarimYas fonksiyonunu çağırır
```

```python
sinifDF.isnull() #NaN değer varmı diye kontrol eder
```

```python
yeniDic={"soyisim":["yıldırım","yıldırım","yıldırım","günes","günes","günes"],
    "isim":["güllü","veli","güllü","banu","irem","burak"],
     "sinif":["A","B","C","D","C","F"],
     "yas":[12,23,24,22,15,19]}
```

```python
yeniDF=pd.DataFrame(yeniDic)
```

 ###### pivot_table verilen indexleri multi index haline getirir ve eğer aynı indexe sahip value varsa default olarak ortalama alır yada "aggfunc" ile ypılacak işleme biz karar veririz (aşağıda "güllü" örnek gösterilebilir)

```python
yeniDF.pivot_table(values=["yas"],index=["soyisim","isim"],aggfunc=np.sum)
```
